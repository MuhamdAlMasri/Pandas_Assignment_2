### Create a function which takes in a set of coordinates (lat and long) and another set, and calculates the distance between them

```python
from math import sin, cos, sqrt, atan2, radians
def distance_measurment(x,y):
    # Approximate radius of earth in km
    R = 6373.0  
    lat1 = x[0] = radians(52.2296756) #lat1
    lon1 = x[0] = radians(21.0122287) #long1
    lat2 = y[0] = radians(52.406374) #lat2
    lon2 = y[1] = radians(16.9251681) #long2

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance

```

### Explanation of .at & .iat

##### They are similar as loc and iloc, but the main difference they allow us to access single value in the table

```python
df_2000.at[5,"price"]
```

```python
df_2000.iat[0,2]
```
